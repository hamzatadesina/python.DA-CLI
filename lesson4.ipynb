## STATISTICAL ANALYSIS IN PYTHON

```python
# 1.CORRELATION
# 2.CHI SQUARE
# 3.ANALYSIS OF VARIANCE
# 4.KRUSKAL WALIS H TEST
```

```python
# correlation. To check if there is any correlaton between columns

rent_df.columns
```

```python
kendal_corr=rent_df[['BHK', 'Rent', 'Size']].corr('kendall')
pearson_corr=rent_df[['BHK', 'Rent', 'Size']].corr('pearson')
spearman_corr=rent_df[['BHK', 'Rent', 'Size']].corr('spearman')

fig, axes= plt.subplots(1,3, figsize= (20,7))

sns.heatmap(data=kendal_corr, ax=axes[0], annot=True, cbar=False)
sns.heatmap(data=pearson_corr, ax=axes[1], annot=True, cbar=False)
sns.heatmap(data=spearman_corr, ax=axes[2], annot=True, cbar=False)

plt.suptitle('CORRELATION MATRIX OF RENT, SIZE AND BHK')

axes[0].set_title('KENDAL-TAU')
axes[1].set_title('PEARSON PRODUCT MOMENT')
axes[2].set_title('SPEARMAN RANK CORRELATION')
```

### Insights

```python
# testing the statistical significance of the correlation

# testing normality -- shapiro wilk/ kolmogrov-smirnov

from scipy.stats import shapiro

Ho = 'There is normalty'
H1 = 'There is no normalty'

col_test = {}
col_list = ['BHK', 'Rent', 'Size']
for col in  col_list:
    stats, p_value= shapiro(rent_df[col].values)
    stats_list = [stats, p_value]
    col_test.update({col: stats_list})

for i, v in col_test.items():
    if v[1] < 0.05:
        print(f'{H1} in {i} @ p_value = {v[1]}')
    else:
        print(f'{Ho} in {i} @ p_value = {v[1]}')
```

```python
stats, p_value = shapiro(rent_df['BHK'].values)

if p_value < 0.05:
    print(f'{H1}')
else:
    print(f'{Ho}')

print(p_value)
```

```python
stats, p_value = shapiro(rent_df['Rent'].values)

if p_value < 0.05:
    print(f'{H1}')
else:
    print(f'{Ho}')

print(p_value)
```

```python
stats, p_value = shapiro(rent_df['Size'].values)

if p_value < 0.05:
    print(f'{H1}')
else:
    print(f'{Ho}')

print(p_value)
```

```python
# testing the statistical significance of Spearman 

from scipy.stats import spearmanr
bhk_rent= stats, p_value= spearmanr(rent_df['BHK'].values, rent_df['Rent'].values)
size_rent= stats, p_value= spearmanr(rent_df['Size'].values, rent_df['Rent'].values)
size_bhk= stats, p_value= spearmanr(rent_df['Size'].values, rent_df['BHK'].values)

test_pairs = {
    'BHK vs Rent': bhk_rent,
    'Size vs Rent': size_rent,
    'size vs bhk': size_bhk

}

for i, v in test_pairs.items():
    if v[1] < 0.05:
        print(f'{H1} in {i} @ p_value = {v[1]} with coeff = {v[0]}')
    else:
        print(f'{Ho} in {i} @ p_value = {v[1]} with coeff = {v[0]}')
```

```python
### Analysis of Variance (ANOVA)... 
# 1. TO CHECK CORRESLATION BETWEEN CATEGORICAL AND CONTINEOUS VARIABLE

Ho = 'There is no statistical significant difference'
H1 = 'There is no statistical significant difference'

from scipy.stats import f_oneway

stats, p_value = f_oneway(rent_df[rent_df['City'] == 'Kolkata']['Size'].values,
                          rent_df[rent_df['City'] == 'Delhi']['Size'].values,
                          rent_df[rent_df['City'] == 'Mumbai']['Size'].values,
                          rent_df[rent_df['City'] == 'Bangalore']['Size'].values,
                          rent_df[rent_df['City'] == 'Chennai']['Size'].values,
                          rent_df[rent_df['City'] == 'Hyderabad']['Size'].values)

if p_value < 0.05:
    print(f'{H1} between sizes in each city @ p_value = {p_value}')
else:
    print(f'{Ho} between sizes in each city @ p_value = {p_value}')

print(p_value)


```
