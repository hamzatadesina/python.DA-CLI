```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
Lauren_df = pd.read_csv('Lauren furniture.csv')
```

```python
#seaborn

# customize seaborn chart

sns.set_style('darkgrid')
sns.set_palette('husl')

sns.barplot(data= Lauren_df, x = 'revenue', y = 'product', hue = 'purchase_size',
            estimator = 'mean')
plt.title('BAR CHART OF AVERAGE REVENUE  BY EACH PRODUCT')



```

```python
# plotly

px.histogram(data_frame= Lauren_df, x = 'revenue', y = 'product', color= 'purchase_size', 
             barmode = 'group', histfunc = 'avg', title = 'BAR CHART OF AVEGARE REVENUE PER PRODUCT')
```
